```python
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy.stats import uniform
plt.rcParams['figure.figsize'] = [8, 8]
```

```python
fig, ax = plt.subplots() # note we must use plt.subplots, not plt.subplot
# (or if you have an existing figure)

board=plt.Rectangle((-1,-1),2,2,edgecolor='black',facecolor='none',alpha=0.5)
circle = plt.Circle((0, 0), 1, edgecolor='r',facecolor='none',alpha=0.6)
x=np.linspace(-1,1,10000)
y=uniform.rvs(-1,2,10000)
ax = plt.gca()
ax.set_xlim((-1.2, 1.2))
ax.set_ylim((-1.2, 1.2))

ax.add_patch(board)
ax.add_patch(circle)
ax.scatter(np.linspace(-1,1,10000),uniform.rvs(-1,2,10000),color='g',alpha=0.1)
ax.scatter(x,y,color='g',alpha=0.1)
```

```python
#distance(radius) of darts from the origin
dist=x**2+y**2
inc=np.where(dist<1)
inp=dist[inc]
p=len(inp)
fraction=p/10000

# Fraction of dart balls within the circle
print(fraction)
```

```python
#For dimennsions 3 to 10
import pandas as pd

data = pd.DataFrame(np.vstack((x, y)).reshape(-1,2), columns=['1','2'])
dim=np.arange(3,11,1)
fract=np.array(fraction)

for i in dim:
  data[i]=uniform.rvs(-1,2,10000)
  fraction=data[((data**2).sum(axis=1)<1)]
  fract=np.append(fract,fraction.shape[0]/10000)

dims=np.arange(2,11,1)
plt.figure(figsize=(10,4))
plt.plot(dims,fract)
plt.title(label='Fraction of dart balls within the circle')
plt.xlabel('Dimension numbers')
plt.ylabel('Dart ball fractions')
```

From the trend it is observed that the  number of dart balls within the circle keeps decreasing as dimension increases.
