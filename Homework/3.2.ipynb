```python
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy.stats import uniform
plt.rcParams['figure.figsize'] = [8, 8]
```

```python
fig, ax = plt.subplots() # note we must use plt.subplots, not plt.subplot
# (or if you have an existing figure)

board=plt.Rectangle((-1,-1),2,2,edgecolor='black',facecolor='none',alpha=0.5)
circle = plt.Circle((0, 0), 1, edgecolor='r',facecolor='none',alpha=0.5)
x=np.linspace(-1,1,10000)
y=uniform.rvs(-1,2,10000)
ax = plt.gca()
ax.set_xlim((-1.2, 1.2))
ax.set_ylim((-1.2, 1.2))

ax.add_patch(board)
ax.add_patch(circle)
#ax.scatter(np.linspace(-1,1,10000),uniform.rvs(-1,2,10000),color='g',alpha=0.1)
ax.scatter(x,y,color='g',alpha=0.1)
```

```python
#distance(radius) of darts from the origin
dist=x**2+y**2
inc=np.where(dist<1)
inp=dist[inc]
p=len(inp)
fraction=p/10000
print(fraction)
```
